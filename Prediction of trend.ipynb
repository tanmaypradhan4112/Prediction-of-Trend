```python
#considering major columns for prediction and analysis.
product_selected_var_df=product.iloc[:,[4,7,8,9,10,11,12]]
```

```python
product_selected_var_df.head()
```

```python
#predicting model through Linear Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x=product_selected_var_df.drop('online_visit',axis=1)
y=product_selected_var_df['online_visit']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.3,random_state=3)
Lin_reg=LinearRegression()
Lin_reg.fit(X_train,Y_train)
Y_pred=Lin_reg.predict(X_test)
plt.scatter(Y_test,Y_pred)
```
